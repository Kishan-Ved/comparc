```python
import matplotlib.pyplot as plt

# Example data
x = [64, 128, 256, 512, 1024]
y = [0.001, 0.002, 0.004, 0.001, 0.007]
y2 = [1.232, 1.256, 1.309, 1.266, 1.301]

# Create the plot for y
plt.plot(x, y, marker='o', label='C++')

# Create the plot for y2
plt.plot(x, y2, marker='s', label='Python')

# Set x-axis ticks
plt.xticks(x)

# Add labels and title
plt.xlabel('Matrix Size (N)')
plt.ylabel('Execution Time (seconds)')
plt.title('System CPU time comparison')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
